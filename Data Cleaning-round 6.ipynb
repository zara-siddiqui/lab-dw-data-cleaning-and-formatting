```python
import pandas as pd
import numpy as np
```

```python
file = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
data = pd.read_csv(file)
display(data.head(3))
data.shape
```

```python
# Exercise 1: Cleaning Column Names
```

```python
for column in data.columns:
    data_columns = data[column].unique()
    print(data_columns)
```

```python
cols = []
for column in data.columns:
    cols.append(column.lower())
cols
data.columns = cols
data
```

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')
data.head(2)
```

```python
data = data.rename(columns = {"st":"state"})
data.head(2)
```

```python
# Exercise 2: Cleaning invalid Values
```

```python
data = data.drop_duplicates(subset=['customer'], keep = False)
print(data.shape)
data.tail(5)
```

```python
data.isna().sum()
```

```python
data['gender'].value_counts(dropna=False)
```

```python
data['gender'] = data['gender'].fillna('F')
data.head(5)
```

```python
gender_column = data['gender'].unique()
print(gender_column)
```

```python
valid_gender = {'F': 'Female', 'Femal': 'Female', 'female' : 'Female', 'M': 'Male'}
data['gender'] = data['gender'].replace(valid_gender)
data.head(5)
```

```python
state_column = data['state'].unique()
print(state_column)
```

```python
valid_state = {'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'}
data['state'] = data['state'].rename(valid_state)
data.head(5)
```

```python
education_column = data['education'].unique()
print(education_column)
```

```python
valid_education = {'Bachelors':'Bachelor'}
data['education'] = data['education'].replace(valid_education)
data.head(5)
```

```python
customer_lifetime_value_column = data['customer_lifetime_value'].unique()
print(customer_lifetime_value_column)
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace('%','')
data['customer_lifetime_value'] = pd.to_numeric(data['customer_lifetime_value'] , errors='coerce')/100
data['customer_lifetime_value']
```

```python
mean = data['customer_lifetime_value'].mean()
mean
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].fillna(mean)
data['customer_lifetime_value']
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].round(decimals = 2)
data['customer_lifetime_value']
```

```python
vehicle_class_column = data['vehicle_class'].unique()
print(vehicle_class_column)
```

```python
valid_vehicle_class={'Luxury SUV' : 'Luxury', 'Sports Car' : 'Luxury' , 'Luxury Car':'Luxury'}
data['vehicle_class'] = data['vehicle_class'].replace(valid_vehicle_class)
data.head(5)
```

```python
vehicle_class_column = data['vehicle_class'].unique()
print(vehicle_class_column)
```

```python
data.isnull().sum()
```

```python
number_of_open_complaints_column = data['number_of_open_complaints'].unique
print(number_of_open_complaints_column)
```

```python
new_list = []
for item in data['number_of_open_complaints']:
    index_item = item.split('/')[1]
    new_list.append(str(index_item))
data['number_of_open_complaints'] = new_list
data.head(5)
```

```python
y = data2['total_claim_amount']
X = data2.drop(['total_claim_amount'], axis=1)

X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
X_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first').fit(X_cat)
cols = encoder.get_feature_names_out(input_features=X_cat.columns)
X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)
display(X_cat_encode)
```

```python
x_normalized_df = pd.DataFrame(x_normalized, columns=['customer_lifetime_value1','income1','monthly_premium_auto1'])
```

```python
display(x_normalized_df)
```

```python
X = pd.concat([x_normalized_df, X_cat_encode], axis=1)
X.head(5)
```

```python
# Linear Regression - Train-Test Split
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression as linreg
```

```python
lm = linreg()
lm.fit(X_train,y_train)
```

```python
# Model Validation
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
r2 = r2_score(y_test, predictions_test)
r2
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
rmse
```
